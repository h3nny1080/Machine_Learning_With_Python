```python
# Mean, Varience and Standard Deviation

x = [30, 45, 60, 50, 40, 55, 70, 35, 65, 50, 42]
total = 0
n = 0
for num in x:
    print(total)
    total = total + num
    n = n + 1
mean = total / n

print(str(mean))
deviations = 0
for num in x:
    deviations = deviations + (num - mean) ** 2

sd = deviations/(n-1)
var = sd**(1/2)

print("mean = "+str(mean))
print("var = "+str(var))
print("sd = "+str(sd))
print("n = " +str(n))

mean =5000
```
